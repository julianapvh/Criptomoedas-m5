## Em quais anos houve quedas e valorizações?

Para realizar esta análise as moedas foram separadas em dois quadrantes contendo cinco em cada, medindo a linha temporal desde o surgimento delas até o marco de 2021.

Esta divisão foi realizada, primeiramente, para facilitar a visualização mais limpa das informações, e depois, para enquadrar melhor o período de surgimento/início da valorização de cada moeda dentro de uma margem.

A criação dos dois gráficos foram feitas com base nos valores registrados em "Alta" disponível na tabela de cada criptomoeda, e para verificar a variação bruta, os valores anuais de cada cripto foram somados, por isso estão na casa dos trilhões de reais. Foi feito a medida por parte destes dados para identificar realmente qual foi o pico máximo que se registraram os valores.

O primeiro gráfico nos mostra o balanço das moedas ChainLink, Cosmos, EOS, Polkadot e USDCoin.
* Da data de criação/lançamento destas criptos, durante seu tempo de vida, nota-se que 2021 foi o ano de maior valor para ChainLink, Polkadot e Cosmos.
* Para a EOS, seu pico foi no ano de 2018 com tendência de congelamento após o ano de 2019
* USDCoin manteve linearidade na estabilidade do seu preço.

Enquanto as demais criptomoedas variam suas tendências, a USDCoin se mantém constante. Isso ocorre porque ela é uma moeda pareada ao dólar só que funcionando em carteiras digitais e com o sistema blockchain, então, técnicamente, seu valor se estabiliza no preço da moeda norte-americana.

A curva da EOS tem alta logo no início de sua operação e depois, uma queda suave com predição de estagnar. Há alguns fatores que podem dar significado a este movimento. A EOS surgiu com um padrão diferenciado de tecnologia blockchain descentralizada, a fácil operação e utilização de tokens (ativos que possuem valor comercial) da moeda gerou bilhões para sua receita.

O movimento de decaída da EOS se deve a problemas internos em relação ao desenvolvimento contínuo da sua ferramenta base (EOSIO), causando falhas e devido a falta de recursos, causou queda na cripto, até 2021, em que surgiu a ENF (EOS Network Foundation) com nova proposta para levantar a moeda.

As criptomedas ChainLink, Polkadot e Cosmos seguem a tendência de subida do gráfico, quase de maneira simultânea. O movimento da Cosmos pode estar ligado ao seu modo de operação, é uma cripto interoperadore, que permite transferencia entre outras moedas, o que pode ter valorizado e permitido a escalabilidade. Já a ChainLink (LINK) e a Polkadot (DOT) iniciaram como plataformas de transação de criptomoedas para superar 'gargalos' deste mercado, como a de criar pontes  entre criptos e a facilidade para os usuários, antes de iniciarem suas operações com moeda própria, o que em si foi um facilitador e assim que suas moedas foram lançadas houve supervalorização.
